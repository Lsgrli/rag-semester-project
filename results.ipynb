```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set_theme(style="whitegrid")
plt.rcParams["figure.dpi"] = 120

%matplotlib inline

print("\n==============================")
print("RAG RESULTS ANALYSIS")
print("==============================")

# Load data
df = pd.read_csv("dataset4_stats.csv")

print("\nDataset loaded")
print("Number of questions:", len(df))

print("\nColumns:")
print(df.columns)

display(df.head())

# GLOBAL METRICS
print("\n==============================")
print("GLOBAL METRICS")
print("==============================")

print("Average LLM score:", round(df["llm_score"].mean(),3))
print("Paper Recall:", round(df["paper_recall"].mean(),3))
print("Chunk Precision:", round(df["chunk_precision"].mean(),3))
print("Top Chunk Accuracy:", round(df["top_chunk_accuracy"].mean(),3))


# MRR (Mean Reciprocal Rank)
def compute_mrr(row):

    target = str(row["target_paper"]).lower().replace(".pdf","")

    retrieved = str(row["retrieved_papers"]).split(";")
    retrieved = [p.lower().replace(".pdf","").strip() for p in retrieved]

    for i, paper in enumerate(retrieved):

        if paper == target:
            return 1/(i+1)

    return 0

df["mrr"] = df.apply(compute_mrr, axis=1)

print("\nMRR =", round(df["mrr"].mean(),3))

# METRICS BY QUESTION TYPE (TAG)
print("\n==============================")
print("METRICS BY QUESTION TYPE")
print("==============================")

tag_stats = df.groupby("tag").agg({
    "llm_score":"mean",
    "paper_recall":"mean",
    "chunk_precision":"mean",
    "top_chunk_accuracy":"mean",
    "mrr":"mean"
}).sort_values("llm_score", ascending=False)

display(tag_stats)

# METRICS BY PAPER
print("\n==============================")
print("METRICS BY PAPER")
print("==============================")

paper_stats = df.groupby("target_paper").agg({
    "llm_score":"mean",
    "paper_recall":"mean",
    "chunk_precision":"mean",
    "top_chunk_accuracy":"mean",
    "mrr":"mean"
})

paper_stats = paper_stats.sort_values("llm_score")

display(paper_stats.head(15))

print("\nWorst papers (retrieval or QA difficulty):")
display(paper_stats.head(10))

print("\nBest papers:")
display(paper_stats.tail(10))

# RETRIEVAL FAILURES
print("\n==============================")
print("RETRIEVAL FAILURES")
print("==============================")

failures = df[df.paper_recall == 0]

print("Number of failures:", len(failures))
print("Failure rate:", round(len(failures)/len(df),3))

display(failures[[
    "question",
    "tag",
    "target_paper",
    "retrieved_papers",
    "llm_score"
]].head(10))


# GENERATION FAILURES
print("\n==============================")
print("GENERATION FAILURES")
print("==============================")

gen_failures = df[(df.paper_recall == 1) & (df.llm_score <= 1)]

print("Generation failures:", len(gen_failures))

display(gen_failures[[
    "question",
    "tag",
    "target_paper",
    "retrieved_papers",
    "llm_score"
]].head(10))

# CHUNK QUALITY ANALYSIS
print("\n==============================")
print("CHUNK QUALITY")
print("==============================")

print("Average chunk precision:", round(df["chunk_precision"].mean(),3))
print("Average top chunk accuracy:", round(df["top_chunk_accuracy"].mean(),3))

print("\nQuestions with very noisy retrieval:")

noisy = df[df.chunk_precision < 0.2]

display(noisy[[
    "question",
    "target_paper",
    "retrieved_papers",
    "chunk_precision"
]].head(10))

# PLOTS
print("\nPlots...")

# MRR distribution

plt.figure(figsize=(6,4))

sns.histplot(df["mrr"], bins=20)

plt.title("MRR Distribution")

plt.show()

# 2 Retrieval vs Generation

plt.figure(figsize=(6,5))

sns.scatterplot(
    x="paper_recall",
    y="llm_score",
    data=df
)

plt.title("Retrieval vs Generation")

plt.xlabel("Paper Recall")
plt.ylabel("LLM Score")

plt.show()

# 3 LLM Score distribution

plt.figure(figsize=(6,4))

sns.histplot(df["llm_score"], bins=[0,1,2,3,4])

plt.title("Distribution of LLM Scores")

plt.show()

# 4 Score by question type

plt.figure(figsize=(8,5))

sns.boxplot(
    x="tag",
    y="llm_score",
    data=df
)

plt.title("LLM Score by Question Type")

plt.show()

# 5 Paper recall by question type

plt.figure(figsize=(8,5))

sns.barplot(
    x="tag",
    y="paper_recall",
    data=df
)

plt.title("Paper Recall by Question Type")

plt.show()

# 6 Chunk precision distribution

plt.figure(figsize=(6,4))

sns.histplot(df["chunk_precision"], bins=20)

plt.title("Chunk Precision Distribution")

plt.show()

# 7 Top chunk accuracy

plt.figure(figsize=(6,4))

sns.histplot(df["top_chunk_accuracy"], bins=2)

plt.title("Top Chunk Accuracy")

plt.show()

# 8 Retrieval vs chunk precision

plt.figure(figsize=(6,5))

sns.scatterplot(
    x="chunk_precision",
    y="llm_score",
    data=df
)

plt.title("Chunk Precision vs LLM Score")

plt.show()


# 9 Tag distribution

plt.figure(figsize=(6,4))

sns.countplot(
    x="tag",
    data=df
)

plt.title("Distribution of Question Types")

plt.show()

# FINAL SUMMARY

print("\n==============================")
print("FINAL DIAGNOSTIC")
print("==============================")

retrieval_failure_rate = len(df[df.paper_recall == 0]) / len(df)

generation_failure_rate = len(df[(df.paper_recall == 1) & (df.llm_score <= 1)]) / len(df)

print("Retrieval failure rate:", round(retrieval_failure_rate,3))
print("Generation failure rate:", round(generation_failure_rate,3))

print("\nInterpretation guide:")
print("- High retrieval failure → improve retriever")
print("- High generation failure → improve prompt or context selection")
print("- Low chunk precision → retrieval noise")
```

```python
# Matrix Retrieval × Generation
matrix = pd.crosstab(
    df["paper_recall"],   # 0 or 1
    df["llm_score"]       # 0–3
)

plt.figure(figsize=(6,4))

sns.heatmap(
    matrix,
    annot=True,
    fmt="d",
    cmap="Blues"
)

plt.title("RAG Error Matrix")
plt.xlabel("LLM Score")
plt.ylabel("Retrieval Correct (Paper Recall)")

plt.show()
```
